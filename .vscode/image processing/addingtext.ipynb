```python
from PIL import Image, ImageDraw,ImageFont
```

```python
plankton=Image.open('pyyy.jpg')
```

```python
drawable=ImageDraw.Draw(plankton)
```

```python
myfont = ImageFont.truetype('font.ttf',30)
```

```python
drawable.text((50,50),"Plankton!The Devil",(225,0,0),font=myfont)
```
